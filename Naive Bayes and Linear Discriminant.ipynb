## Naive Bayes Algorithm

```python
df.drop(['workclass','education','marital-status','occupation','relationship','race','native-country','income'],axis='columns',inplace=True)
df.head()
```

```python
x=df.drop('age',axis='columns')
target=df.age
```

```python
y=pd.get_dummies(x.gender)
y.head(5)
```

```python
x=pd.concat([x,y],axis='columns')
x.head(5)
```

```python
x.drop(['gender','Male'],axis='columns',inplace=True)
x.head(5)
```

```python
x.columns[x.isna().any()]
```

```python
x.fnlwgt[:5]
```

```python
x.fnlwgt=x.fnlwgt.fillna(x.fnlwgt.mean())
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,target,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(),X_train, y_train, cv=5)
```

## Linear Discriminant Algorithm

```python
lda = LinearDiscriminantAnalysis()
lda.fit(X_train,y_train)
predictions_lda = lda.predict(X_test)
print("Accuracy Score is:")
print(accuracy_score(y_test, predictions_lda))
```

## Conclusion:

NBA and LDA are both linear classifiers. From the observations, we can see that LDA has more accuracy than NBA because LDA takes the covariance of the variables in the dataset.
